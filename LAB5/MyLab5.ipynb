```python
import pandas as pd
```

```python
df1=pd.DataFrame({"key":[1,2,3,4], "value":['A','B','C','D']})
df2=pd.DataFrame({"key":[1,2,3,4], "value":['E','F','G','H']})
display(df1,df2)
```

```python
df1.columns=('value', 'key')
display(df1)
```

```python
df1.drop('value', axis=1, inplace=True)
display(df1)
```

```python
df1.drop(1, axis=0, inplace=True)
display(df1)
```

```python
concat=pd.concat([df1,df2], ignore_index=True)
concat
```
